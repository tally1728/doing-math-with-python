### \#2: Exploring a Quadratic Function Visually

2次関数 $f(x) = x^2 + 2x + 1$ のグラフ

```python
import matplotlib.pyplot as plt


def quad(x):
    return x ** 2 + 2 * x + 1


x = range(-10, 10)
y = [quad(a) for a in x]

fig, ax = plt.subplots()
ax.set_title("Quadratic Function")
ax.plot(x, y)
ax.hlines(0, xmin=-10, xmax=7.5, colors="g", linestyles="dashed")
plt.show()
```
