### \#2: Law of Large Numbers

サイコロを振る回数を徐々に増やして、出る目の平均値が期待値に近づくことを確かめる

```python
import random

import matplotlib.pyplot as plt


def trials(n):
    return sum([random.randint(1, 6) for i in range(n)]) / n
```

まずは1回から10回まで。

```python
# 1 - 10 times
n_trials = list(range(1, 11))
averages = [trials(n) for n in n_trials]
for n, avg in zip(n_trials, averages):
    print("Trials: {0:,} Trial average {1:,.5f}".format(n, avg))

fig, ax = plt.subplots()
ax.scatter(n_trials, averages)
ax.hlines(3.5, xmin=1, xmax=10, colors="g", linestyles="dashed")
ax.set_title("The Average Value of Results & The Expectation")
ax.set_xlabel("The Number of Trials")
ax.set_xscale("log")
ax.set_ylabel("The Average Value of Results")
fig.savefig("chap5_pc2_law-of-large-numbers_10times.png")
plt.show()
```

次に10回から500,000回まで。

```python
n_trials = [10, 100, 1_000, 10_000, 100_000, 500_000]
averages = [trials(n) for n in n_trials]
for n, avg in zip(n_trials, averages):
    print(
        "Trials: {0:,} Trial average {1:,.5f} (Difference: {2:.5f})".format(
            n, avg, avg - 3.5
        )
    )

fig, ax = plt.subplots()
ax.scatter(n_trials, averages)
ax.hlines(3.5, xmin=10, xmax=500_000, colors="g", linestyles="dashed")
ax.set_title("The Average Value of Results & The Expectation")
ax.set_xlabel("The Number of Trials (Log Scale)")
ax.set_xscale("log")
ax.set_ylabel("The Average Value of Results")
fig.savefig("chap5_pc2_law-of-large-numbers_500ktimes.png")
plt.show()
```
