### \#1: Packing Circles into a Square

[matplotlib.patches.Polygon](https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.Polygon.html?highlight=polygon#matplotlib.patches.Polygon) クラスで正方形を描く

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_xlim(left=0, right=6)
ax.set_ylim(bottom=0, top=6)
ax.set_aspect("equal")

square = patches.Polygon([(1, 1), (5, 1), (5, 5), (1, 5)], closed=True)
ax.add_patch(square)

plt.show()
```

パッキング問題（四角形に円を出来る限り詰め込む）の簡易版

```python
import matplotlib.patches as patches
import matplotlib.pyplot as plt

# the Size of Square
size = (4, 4)

fig, ax = plt.subplots()
ax.set_xlim(left=0, right=size[0])
ax.set_ylim(bottom=0, top=size[1])
ax.set_aspect("equal")

y = 0.5
while y < size[1]:
    x = 0.5
    while x < size[0]:
        c = patches.Circle((x, y), radius=0.5)
        ax.add_patch(c)

        x += 1
    y += 1

plt.show()
```
