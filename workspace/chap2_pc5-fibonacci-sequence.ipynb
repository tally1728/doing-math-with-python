### \#5: Exploring the Relationship Between the Fibonacci Sequence and the Golden Ratio

```python
import matplotlib.pyplot as plt


def fibo(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1, 1]

    # n> 2
    a = 1
    b = 1
    # First two members of the series
    series = [a, b]
    for i in range(n):
        c = a + b
        series.append(c)
        a = b
        b = c

    return series


fig, ax = plt.subplots()

ax.set_title("Ratio between consecutive Fibonacci numbers")
ax.set_xlabel("index of the Fibonacci Sequence")
ax.set_ylabel("Ratio")

fibo_seq = fibo(10)
ratio = [n / m for m, n in zip(fibo_seq, fibo_seq[1:])]
ax.plot(ratio)
ax.hlines(1.6180339887, xmin=0, xmax=10, colors="g", linestyles="dashed")

plt.show()
```
