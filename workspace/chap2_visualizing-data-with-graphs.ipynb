### Comparing the Monthly Temperature Trends of New York City

```python
nyc_temp_2000 = [31.3, 37.3, 47.2, 51.0, 63.5, 71.3, 72.3, 72.7, 66.0, 57.0, 45.3, 31.1]
nyc_temp_2006 = [40.9, 35.7, 43.1, 55.7, 63.1, 71.0, 77.9, 75.8, 66.6, 56.2, 51.9, 43.6]
nyc_temp_2012 = [37.3, 40.9, 50.9, 54.8, 65.1, 71.0, 78.8, 76.7, 68.8, 58.0, 43.9, 41.5]

months = range(1, 13)

fig, ax = plt.subplots()
ax.plot(months, nyc_temp_2000, marker="o")
ax.plot(months, nyc_temp_2006, marker="o")
ax.plot(months, nyc_temp_2012, marker="o")
ax.legend([2000, 2006, 2012])
plt.show()
```

### Customizing Graphs

#### Adding a Title and Labels

```python
fig, ax = plt.subplots()
ax.set_title("Average monthly temperature in NYC")
ax.set_xlabel("Month")
ax.set_ylabel("Temperature")
ax.plot(months, nyc_temp_2000, marker="o")
ax.plot(months, nyc_temp_2006, marker="o")
ax.plot(months, nyc_temp_2012, marker="o")
ax.legend([2000, 2006, 2012])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(years, nyc_temp, marker="o")
ax.set_ylim(bottom=40, top=60)
plt.show()
```

### Plotting with Formulas

#### Newton’s Law of Universal Gravitation

```python
"""
The relationship between gravitational force and distance between two bodies
"""


# Draw the graph
def draw_graph(x, y):
    fig, ax = plt.subplots()
    ax.plot(x, y, marker="o")
    ax.set_title("Gravitational force and distance")
    ax.set_xlabel("Distance in meters")
    ax.set_ylabel("Gravitational force in newtons")
    plt.show()


# Calc gravitational force
def calc_force(m1, m2, d):
    G = 6.674 * (10 ** -11)
    return G * (m1 * m2) / (d ** 2)


def generate_F_r():
    # Two masses
    m1 = 0.5
    m2 = 1.5

    # Generate values for r
    r = range(100, 1001, 50)
    # Empty list to store the calculated values of F
    F = []

    # Calculate force and add it to the list, F
    for dist in r:
        F.append(calc_force(m1, m2, dist))

    # Call the draw_graph function
    draw_graph(r, F)


generate_F_r()
```

#### Projectile Motion

```python
import math


# Generate equally spaced floating point numbers between two given values
def f_range(start, final, increment):
    numbers = []

    while start < final:
        numbers.append(start)
        start += increment

    return numbers


def calc_trajectory(u, theta):
    g = 9.8

    theta = math.radians(theta)

    # Time of flight
    t_flight = 2 * u * math.sin(theta) / g
    # time intervals
    intervals = f_range(0, t_flight, 0.001)

    # List of x and y coordinates
    x = []
    y = []
    for t in intervals:
        x.append(u * math.cos(theta) * t)
        y.append(u * math.sin(theta) * t - 0.5 * g * (t ** 2))

    return (x, y)


def draw_graph():
    fig, ax = plt.subplots()

    ax.set_title("Projectile motion of a ball")
    ax.set_xlabel("x-coordinate (Distance)")
    ax.set_ylabel("y-coordinate (Height)")

    for angle in range(25, 75, 10):
        x, y = calc_trajectory(25, angle)
        ax.plot(x, y, label="Angle {0} deg".format(angle))

    ax.legend()
    plt.show()
    
    fig.savefig("chap2_visualizing-data-with-graphs_projectile-motion.png")


draw_graph()
```
