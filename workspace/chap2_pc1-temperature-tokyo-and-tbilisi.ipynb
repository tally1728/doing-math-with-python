### \#1: How Does the Temperature Vary During the Day?

東京とトビリシの気温変化を比較

```python
# Hourly Temperature at 2021-12-19
tokyo = [3, 2, 2, 2, 1, 2, 3, 2, 3, 4, 6, 7, 8, 8, 9, 9, 8, 7, 6, 6, 5, 4, 4, 3]
tbilisi = [6, 5, 4, 4, 4, 3, 2, 3, 3, 4, 5, 6, 7, 8, 8, 10, 9, 8, 7, 6, 5, 5, 4, 4]

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.set_title("Hourly Temperature at 2021-12-19")
ax.set_xlabel("Local Time")
ax.set_ylabel("Temperature")
ax.plot(tokyo, label="Tokyo")
ax.plot(tbilisi, label="Tbilisi")
ax.legend()
plt.show()
```
