### \#2: Graphical Equation Solver

```python
import sympy

x, y = sympy.symbols("x, y")

eq1 = 2 * x - 3 * y + 2
eq2 = x + 2 * y - 6

sympy.plotting.plot(
    sympy.solve(eq1, y)[0],
    sympy.solve(eq2, y)[0],
    legend=True,
)
```

```python
sympy.solve((eq1, eq2))
```
