### \#2: Drawing the Sierpi≈Ñski Triangle

```python
import random

import matplotlib.pyplot as plt


def transformation_1(p):
    x = 0.5 * p[0]
    y = 0.5 * p[1]
    return x, y


def transformation_2(p):
    x = 0.5 * p[0] + 0.5
    y = 0.5 * p[1] + 0.5
    return x, y


def transformation_3(p):
    x = 0.5 * p[0] + 1
    y = 0.5 * p[1]
    return x, y


def transform(p):
    # List of transformation functions
    transformations = [
        transformation_1,
        transformation_2,
        transformation_3,
    ]
    # Pick a random transformation function and call it
    t = random.choice(transformations)
    return t(p)


# n: the number of iterations
def generate(n):
    p = (0, 0)

    x = [p[0]]
    y = [p[1]]

    for i in range(n):
        p = transform(p)
        x.append(p[0])
        y.append(p[1])
    return x, y


fig, axes = plt.subplots(1, 2, figsize=(15, 7))
fig.suptitle("Sierpi≈Ñski Triangle")
axes = axes.flatten()

for i, n in enumerate([1000, 10_000]):
    x, y = generate(n)

    axes[i].plot(x, y, "o", alpha=0.3)

    axes[i].set_title("{0:,} Iterations".format(n))
    axes[i].set_xlabel("X")
    axes[i].set_ylabel("Y")

fig.savefig("chap6_pc2_drawing-the-sierpinski-triangle.png")
plt.show()
```
